## YARN - Yet another resource negotiatior

This is compared to spark standalone

Its an execution system for hadoop. Most HDFS has YARN configured. 
- You dont have to install spark on all nodes in the cluster. 
- You can run types of java applications, not just spark (default was only spark)
- Can isolate and priority applications (management)
- its the only that supports kerberos-secured HDFS (IAM)
- Runs in a container - JVM - java virtual machine
- YARN supports Hadoop, Spark doesn't
- Takes longer time, because you are creating containers JVM that take more time


Analogy to the standalone spark items

- Resource manager ( similar to the spakrs master process)
- Node Manager similar to worker process
- Application mastetr ( similar to cluster manager)

#### Driver / executor / Master / Cluster manager

Slightly different terminology. Client will launch the driver. Driver will talk to other ocmputers to launch the executor. The executor will run and come back.

## Example 2
- import supervisior filter 94103 and save under supervisor 94103

The number of files created when saving, is dependent on the number of partitions. You should have a single file for each partition that you have. Though some of the files might be empty, since there is no data.

````python
tf = sc.textFile('../../github_repos/2017-msan694-example/Data/supervisor_sf.csv')
tf = tf.map(lambda x : x.split(','))
tf = tf.map(lambda x: (x[0],x[1]))
tf = tf.filter(lambda x : x[0]=='94103')
tf.saveAsTextFile('supervisor_94103')
#tf.collect()
````
